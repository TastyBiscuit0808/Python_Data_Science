# Probability Distributions

A probability distribution describes how a random variable is distributed; it tells us which values a random variable is most likely to take on and which values are less likely.


In this tutorial we'll learn about some of the popular probability distributions:
* Normal distribution
* Bernoulli distribution
* Binomial distibution
* Poisson distribution

```python
# Dependencies

# Standard Dependencies
import numpy as np
import pandas as pd

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Statistics
import scipy.stats as stats
from scipy.stats import binom, poisson, bernoulli
```

### Normal Distribution

Normal distribution, also known as the Gaussian distribution, is a probability distribution that is symmetric about the mean, showing that data near the mean are more frequent in occurrence than data far from the mean.

```python
# Generate Normal Distribution
normal_dist = np.random.randn(10000)
normal_df = pd.DataFrame({'value' : normal_dist})
```

```python
# Scatterplot
plt.figure(figsize=(18,5))
sns.scatterplot(data=normal_df)
plt.legend(fontsize='xx-large')
plt.title('Scatterplot of a Normal Distribution', fontsize='xx-large')
```

```python
# Normal Distribution as a Bell Curve
plt.figure(figsize=(18,5))
sns.distplot(normal_df)
plt.title('Normal distribution (n=1000)', fontsize='xx-large')
```

```python
print("Mean:", round(normal_df.mean()[0]))
print("Median:", round(normal_df.median()[0]))
print("Variance:", round(normal_df.var()[0]))
```

### Bernoulli Distribution


A Bernoulli distribution is the probability distribution for a series of Bernoulli trials where there are only two possible outcomes.
 

```python
# Chance of heads (outcome 1)
p = 0.6

# Create Bernoulli samples
bern_dist = stats.bernoulli.rvs(p, size=1000)
bern_df = pd.DataFrame({'value' : bern_dist})
bern_values = bern_df['value'].value_counts()

# Plot Distribution
plt.figure(figsize=(18,5))
bern_values.plot(kind='bar', rot=0)
plt.annotate(xy=(0.85,300), 
             s='Samples that came up Tails\nn = {}'.format(bern_values[0]), 
             fontsize='large', 
             color='white')
plt.annotate(xy=(-0.2,300), 
             s='Samples that came up Heads\nn = {}'.format(bern_values[1]), 
             fontsize='large', 
             color='white')
plt.title('Bernoulli Distribution: p = 0.6, n = 1000')
```

```python
print("Mean:", round(bern_df.mean()[0],2)) # =p
print("Variance:", round(bern_df.var()[0],2)) # =p*(1-p)
```

### Binomial Distribution

The binomial distibution is a collection of Bernoulli trials for the same event, i.e., it contains more than 1 Bernoulli event for the same scenario for which the Bernoulli trial is calculated. 

```python
n=100
p=0.6
x = np.random.binomial(n, p, size=1000)
plt.figure(figsize=(18,5))
sns.distplot(x, hist=True, kde=True)
```

```python
mean, var = binom.stats(n, p)
print("Mean:", mean) # =n*p
print("Variance:", var) # =n*p*(1-p)
```

### Poisson Distribution

The Poisson distribution models the probability of seeing a certain number of successes within a time interval, where the time it takes for the next success is modeled by an exponential distribution.

```python
#generate Poisson distribution with sample size 10000
mu=6
x = poisson.rvs(mu, size=10000)
plt.hist(x, density=True, edgecolor='black')
plt.title('Poisson distribution with mu=6', fontsize='x-large')
plt.xlabel('x')
plt.ylabel('probability of x')
plt.show()
```

```python
mean, var = poisson.stats(mu)
print("Mean:", mean) # =mu
print("Variance:", var) # =mu
```
